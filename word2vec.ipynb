**WORD ANALOGIES** - In word analogies task we complete the sentence (if a is to b then c is to __). An example is if a man is related to women the a king is to ?. in this we are finding d such tha all are related to each other by the equation **b-a=d-c**

```
import gensim
from gensim.models import word2vec
from gensim.models import KeyedVectors
from sklearn.metrics.pairwise import cosine_similarity
import nltk
from nltk.corpus import stopwords
```

```
nltk.download("punkt")
nltk.download('stopwords')
```

```
stop=set(stopwords.words('english'))

def readfile(file):
  f = open(file,'r',encoding='utf-8')
  text = f.read()

  sentence= nltk.sent_tokenize(text)
  data =[]
  for sent in sentence:
    word=nltk.word_tokenize(sent)
    word=[w.lower() for w in word if len(w)>2 and w not in stop]
    data.append(word)
  
  return data

text = readfile("/content/notes.txt")
print(text)
```

```
from gensim.models import Word2Vec
model = Word2Vec(text,size=300,window=10,min_count=1)
print(model)
```

```
words=list(model.wv.vocab)
print(words)
```

```
print(model["deepika"])
print(model["deepika"].shape)
```

```
def predict(a,b,c,word_vectors):
  a,b,c = a.lower(),b.lower(),c.lower()
  max_similarity= -100
  d=None

  wa,wb,wc = word_vectors[a], word_vectors[b], word_vectors[c]
  options=['deepika','deepika','singh','padukone','priyanka','nick','jones','virat','kohli','anushka','sharma']

  for w in options:
    if w in [a,b,c]:
      continue
    
    wv = word_vectors[w]
    sim = cosine_similarity([wb-wa],[wv-wc])
    if sim> max_similarity:
      max_similarity=sim
      d = w
    
    return d
```

```
triad=('nick','priyanka','ranveer')
x = predict(*triad,model.wv)
```
